# Training

## Training (part 1)

```python
run_name = "008" # changed resized to PIL.Image.Bilinear
run_id = None
```

```python
# Check the configuration of the experiment
config_exp
```

```python
!nvidia-smi
```

```python
# Treinamento
compare=float('inf')

model = EncoderDecoder(pretrained_model = (config_exp["encoder_pretrained_model"], config_exp["decoder_pretrained_model"]), 
                       eos_token_id=tokenizer.eos_token_id, 
                       pad_token_id=tokenizer.pad_token_id)

model = model.to(device)
torch.cuda.empty_cache()
# DataLoader
train_loader = DataLoader(train_dataset, batch_size=config_exp["batch_size"], shuffle=True, collate_fn = collate_custom, drop_last=True,  num_workers=3, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=config_exp["batch_size"], collate_fn = collate_custom, drop_last=True, num_workers=3, pin_memory=True)

# Optmizer
optimizer = torch.optim.Adam(model.parameters(), lr=config_exp["learning_rate"])


# Starting wandb
if run_id is not None:
    print(run_id)
    run = wandb.init(project="ImageCaptioning_Project", name=f"experiment_{run_name}", config=config_exp, resume=run_id, id=run_id, settings=wandb.Settings(start_method="thread"), reinit=True, dir=os.getenv("WANDB_DIR", config_exp["path_save_checkpoints"]))
    
    if wandb.run.resumed:
        print("Wandb --resumed")
        wandb.restore(config_exp["path_save_checkpoints"]+config_exp["checkpoint_folder_name"])
        checkpoint = torch.load(config_exp["path_save_checkpoints"]+config_exp["checkpoint_folder_name"])
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        epoch_init = checkpoint['epoch']+1
        loss = checkpoint['loss']
else:
    epoch_init = 0
    run = wandb.init(project="ImageCaptioning_Project", name=f"experiment_{run_name}", config=config_exp, resume=True, settings=wandb.Settings(start_method="thread"), reinit=True, dir=os.getenv("WANDB_DIR", config_exp["path_save_checkpoints"]))
    
# Staring Loop of epochs
for epoch in range(epoch_init, config_exp["epochs"]):
    torch.cuda.empty_cache()
    train_loss_batch = []
    val_loss_batch = []

    # Training Loop
    for batch, (train_inputs, train_targets, train_all_targets, train_ids) in enumerate(tqdm(train_loader)):
        model.train()
        model.zero_grad()

        train_inputs = train_inputs.to(device)
        train_targets = train_targets.to(device)

        # Applying mask of value -100  for all tokens =  [PAD]
        mylabels = train_targets.clone()
        mylabels[mylabels==tokenizer.pad_token_id] = -100

        train_output = model(images = train_inputs, targets=mylabels)
        logits = train_output.logits
        loss = train_output.loss

        loss.backward()
        optimizer.step()

        train_loss_batch.append(loss.item())
        
        wandb.log({"train/batch": batch,
                   "train/batch/loss": loss.item()})
        
    train_loss_epoch = sum(train_loss_batch)/len(train_loss_batch)


    # Validation Loop
    with torch.no_grad():
        bleu_list1, bleu_list2, bleu_list3, bleu_list4 = [], [], [], []
        for val_batch, (val_inputs, val_targets, val_all_targets, val_ids) in enumerate(tqdm(val_loader)):
            val_inputs = val_inputs.to(device)
            val_targets = val_targets.to(device)

            model.eval()

            # Model inference in the validation dataset
            labels_val = val_targets.clone()
            labels_val[labels_val==tokenizer.pad_token_id] = -100
            val_output = model(images = val_inputs, targets=labels_val)

            val_loss_batch.append(val_output.loss.item())

            wandb.log({"val/batch": val_batch, 
                    "val/batch/loss": val_output.loss.item()})
            
            if epoch % config_exp["eval_epochs"] == 0:
                decoded_ids = torch.full((val_inputs.shape[0], 1),
                                        model.decoder_start_token_id,
                                        dtype=torch.long).to(val_inputs.device)

                for step in range(config_exp["max_seq_length"]):
                    output_val =  model(images =  val_inputs, decoder_ids= decoded_ids)
                    
                    val_logits = output_val.logits
                    next_token_logits = val_logits[:, -1, :]
                    next_token_id = next_token_logits.argmax(1).unsqueeze(-1)
                    decoded_ids = torch.cat([decoded_ids, next_token_id], dim=-1)

                    # Check if reached end of senquence for all sentences in the batch
                    if torch.eq(next_token_id[:, -1], model.eos_token_id).all():
                        break

                val_preds_sentences = tokenizer.batch_decode(decoded_ids,  skip_special_tokens=True)
                val_preds_sentences = list(map(str2,val_preds_sentences))
                bleu1, bleu2, bleu3, bleu4 = calculate_metrics(val_preds_sentences, val_all_targets)
        
                bleu_list1.append(bleu1)
                bleu_list2.append(bleu2)
                bleu_list3.append(bleu3)
                bleu_list4.append(bleu4)

                wandb.log({"val/batch/BLEU-1":bleu1,
                        "val/batch/BLEU-2":bleu2,
                        "val/batch/BLEU-3":bleu3,
                        "val/batch/BLEU-4":bleu4})

    val_loss_epoch = sum(val_loss_batch)/len(val_loss_batch)
    

    wandb.log({"epoch": epoch, 
            "train/epoch/loss": train_loss_epoch,
            "val/epoch/loss": val_loss_epoch})   
    
    if val_loss_epoch<compare:
        compare=val_loss_epoch
        torch.save(model, config_exp["path_save_model"])

    if epoch % config_exp["eval_epochs"] == 0:
        val_bleu1 = sum(bleu_list1)/len(bleu_list1)
        val_bleu2 = sum(bleu_list2)/len(bleu_list2)
        val_bleu3 = sum(bleu_list3)/len(bleu_list3)
        val_bleu4 = sum(bleu_list4)/len(bleu_list4)

        plt.figure()
        plt.imshow(unorm(val_inputs[-1].cpu()).permute(1,2,0))
        plt.title("Predição: "+str(val_preds_sentences[-1] + " ---- GT: " +str(val_all_targets[-1][0])))
        plt.show()
        plt.pause(0.05)
        print(f'Época: {epoch:d}/{config_exp["epochs"] - 1:d} Train Loss: {train_loss_epoch:.6f} Valid Loss: {val_loss_epoch:.6f}/ Valid BLEU-4: {val_bleu4:.2f}')

        wandb.log({"val/epoch/bleu_score1": val_bleu1,
                   "val/epoch/bleu_score2": val_bleu2,
                   "val/epoch/bleu_score3": val_bleu3,
                   "val/epoch/bleu_score4": val_bleu4})

    else:        
        print(f'Época: {epoch:d}/{config_exp["epochs"] - 1:d} Train Loss: {train_loss_epoch:.6f} Valid Loss: {val_loss_epoch:.6f}')

    # Saving Checkpoint
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': loss,
        }, config_exp["path_save_checkpoints"]+config_exp["checkpoint_folder_name"])
    
    wandb.save()
```
